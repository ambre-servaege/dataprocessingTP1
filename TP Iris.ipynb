```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

#configuration d'affichage
%matplotlib inline
sns.set(style="whitegrid")
```

```python
df=pd.read_csv('/Users/ambreservaege/Downloads/Iris.csv')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Species', y='Id', data =df)
plt.title('Boxplot du prix du billet par classe')
plt.show()
```

```python
df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].hist(bins=20)
plt.show()
```

```python
sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.title("Distribution de la longueur du spale par espce")
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title("Distribution de la largeur du spale par espce")
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='PetalLengthCm', data=df)
plt.title("Distribution de la longueur du ptale par espce")
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='PetalWidthCm', data=df)
plt.title("Distribution de la largeur du ptale par espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='PetalLengthCm', y='Species')
plt.title("relation entre la longueur et l'espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='PetalWidthCm', y='Species')
plt.title("relation entre la largeur et l'espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='SepalLengthCm', y='Species', data =df)
plt.title("Boxplot de la longueur Sepal en fonction de l'espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='PetalLengthCm', y='Species', data =df)
plt.title("Boxplot de la longueur Sepal en fonction de l'espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='SepalWidthCm', y='Species', data =df)
plt.title("Boxplot de la longueur Sepal en fonction de l'espce")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='PetalWidthCm', y='Species', data =df)
plt.title("Boxplot de la longueur Sepal en fonction de l'espce")
plt.show()
```

```python
corr_matrix=df.corr(numeric_only=True)

#affichage de la matrice 
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de corrlation')
plt.show()
```
