## 1. Chargement des données
- Utilisez `pandas` pour charger le fichier CSV `student_performance.csv`.
- Affichez les premières lignes du dataset.



```python
# Espace pour votre code ici

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 

#configuration d'affichage
%matplotlib inline
sns.set(style="whitegrid")

# Charger les données
df=pd.read_csv('/Users/ambreservaege/Downloads/student_performance - student_performance.csv')
df.head()
```


## 2. Relation entre heures d'étude et scores
- Utilisez un **scatter plot** pour visualiser la relation entre les heures d'étude (`Hours_Studied`) et les scores en mathématiques (`Math_Score`).



```python
# Espace pour votre code ici
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Hours_Studied', y='Math_Score')
plt.title("relation entre les heures d'étude et les scores en mathématiques")
plt.show()
# Visualisation de la relation heures d'étude et scores
```


## 3. Corrélations entre les scores
- Calculez la corrélation entre les scores en mathématiques et en lecture.
- Utilisez une matrice de corrélation pour explorer les relations entre les variables numériques.



```python
# Espace pour votre code ici

corr_matrix=df.corr(numeric_only=True)

#affichage de la matrice 
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()
```


## 4. Réduction de dimension avec PCA
- Normalisez les colonnes `Math_Score` et `Reading_Score`.
- Effectuez une réduction de dimension avec l'analyse en composantes principales (PCA).
- Visualisez les résultats sous forme de graphique.



```python
# Espace pour votre code ici

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# PCA et visualisation
df_numeric = df[['Math_Score','Reading_Score']].dropna()
df_scaled=(df_numeric - df_numeric.mean())/df_numeric.std()

pca=PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)

plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1])
plt.title("PCA sur MAth_score et Reading_score")
plt.show()
```
